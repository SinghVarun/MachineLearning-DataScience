```python
%matplotlib notebook
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
np.set_printoptions(2)
```

```python
fruits = pd.read_table("fruit_data_with_colors.txt")
feature_names_fruits = ["height","width", "mass","color_score"]
X_fruits = fruits[feature_names_fruits]
y_fruits = fruits['fruit_label']
target_names_fruits = ['apple', 'mandarin', 'orange', 'lemon']
```

```python
X_fruits_2d = fruits[['height', 'width']]
y_fruits_2d = fruits['fruit_label']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_fruits, y_fruits, random_state=0)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
# Scale the each individual feature on default scale of 0-1
# this is done so that data is centered around the mean of 0
# and has a predictable unit variance.
X_train_scaled = scaler.fit_transform(X_train)
# we must apply the scaling to the test set that we computed for the training set
X_test_scaled = scaler.transform(X_test)
knn = KNeighborsClassifier(n_neighbors = 5);
knn.fit(X_train_scaled, y_train);

print('Accuracy of K-NN classifier on training set: {:.2f}'
     .format(knn.score(X_train_scaled, y_train)))

print('Accuracy of K-NN classifier on test set: {:.2f}'
     .format(knn.score(X_test_scaled, y_test)))


example_fruit = [[5.5, 2.2, 10, 0.70]]
## Because fruit label starts with 1 and ends with 4
target_names_fruits[knn.predict(example_fruit)[0] - 1]
```

```python
cmap_bold = ListedColormap(['#FFFF00', '#00FF00', '#0000FF','#000000'])
from sklearn.datasets import make_regression
plt.figure()

plt.title('Sample regression problem with one input variable')
## Leverage random liner regression.
# n_informative variable indicates number of feature which will used.
X_R1, y_R1 = make_regression(n_samples = 100, n_features=1, n_informative=1, 
                            bias=150.0, noise=30, random_state=0)
plt.scatter(X_R1, y_R1, marker= 'o', s=50)
plt.show()
```

```python
# synthetic dataset for more complex regression
from sklearn.datasets import make_friedman1
plt.figure()
plt.title('Complex regression problem with one variable')
X_F1, y_F1 = make_friedman1(n_samples = 100, n_features = 7, random_state = 0)
plt.scatter(X_F1[:, 2], y_F1, marker='o', s=50, cmap=cmap_bold)
plt.show()


# synthetic dataset for classification (binary) 
plt.figure()
plt.title('Sample binary classification problem with 2 informative features')
# n_informative feature = 2 (as its binary)
# n_feature = sum(n_informative, n_redundant, n_repeated)
X_C2, y_C2 = make_classification(n_samples = 100, n_features = 2, n_redundant=0, 
                                n_informative=2, n_clusters_per_class=1, 
                                flip_y=0.1, class_sep =0.5, random_state=0)

plt.scatter(X_C2[:, 0], X_C2[:, 1], c= y_C2, marker='o', s=50, cmap=cmap_bold)
plt.show()
```

```python
fig, subaxes = plt.subplots(1, 3, figsize=(8, 4))
X_predict_input = np.linspace(-3, 3, 50).reshape(-1, 1)
X_train, X_test, y_train, y_test = train_test_split(X_R1[0::5], y_R1[0::5], random_state = 0)

for thisaxis, K in zip(subaxes, [1, 3, 15]):
    knnreg = KNeighborsRegressor(n_neighbors = K).fit(X_train, y_train)
    y_predict_output = knnreg.predict(X_predict_input)
    thisaxis.plot(X_predict_input, y_predict_output, '^', markersize=10, label='Predicted', alpha=0.8)
    thisaxis.plot(X_train, y_train, 'o', label='True Label', alpha=0.8)
    thisaxis.set_xlabel('Input feature')
    thisaxis.set_ylabel('Target value')
    thisaxis.set_title('KNN regression (K={})'.format(K))
    thisaxis.legend()
    plt.tight_layout()
```

Regression model complexity as a function of K

```python
# plot k-NN regression on sample dataset for different values of K

fig, ax = plt.subplots(5, 1, figsize=(5, 20))
X_predict_input = np.linspace(-3, 3, 50).reshape(-1, 1)
X_train, X_test, y_train, y_test = train_test_split(X_R1, y_R1, random_state=0)


for thisaxis, K in zip(ax, [1, 3, 7, 15, 55]):
    knnReg = KNeighborsRegressor(n_neighbors = K)
    knnReg.fit(X_train, y_train)
    y_predict_output = knnReg.predict(X_predict_input)
    train_score = knnReg.score(X_train, y_train)
    test_score = knnReg.score(X_test, y_test)
    thisaxis.plot(X_predict_input, y_predict_output)
    thisaxis.plot(X_train, y_train, 'o', alpha=0.9, label='Train')
    thisaxis.plot(X_test, y_test, '^', alpha=0.9, label='Test')
    thisaxis.set_xlabel('Input feature')
    thisaxis.set_xlabel('Target value')
    thisaxis.set_title('KNN Regression (K={})\nTrain $R^2 = {:.3f}$, Test $R^2 = {:.3f}$'.format(K, train_score, test_score))
    thisaxis.legend()
    plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=1.0)

```

### Ridge Regression

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_Crime, y_Crime, random_state=0)
# Here because we havent normalize the features, the test scores 
# doesnt look very great.
linridge = Ridge(alpha=20).fit(X_train, y_train)
print("Crime dataset")
print("Ridge regression linear model intercept: {}".format(linridge.intercept_))
print("Ridge regression linear model coefficient: {}".format(linridge.coef_))
print("R-Squared score (training) : {:.3f}".format(linridge.score(X_train, y_train)))
print("R-Squared score (test) : {:.3f}".format(linridge.score(X_test, y_test)))
print("Number of non-zero features {}".format(np.sum(linridge.coef_ != 0)))
```

### ridge regression with feature normalization

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train, X_test, y_train, y_test = train_test_split(X_Crime, y_Crime, random_state = 0)

## Remember you dont fit (X_test) data, else this will lead to data leakage 
# Resulting in high predictibility arising 
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

linridge = Ridge(alpha = 20).fit(X_train_scaled, y_train)
print("Crime dataset")
print("Ridge regression linear model intercept: {}".format(linridge.intercept_))
print("Ridge regression linear model coefficient: {}".format(linridge.coef_))
print("R-Squared score (training) : {:.3f}".format(linridge.score(X_train_scaled, y_train)))
print("R-Squared score (test) : {:.3f}".format(linridge.score(X_test_scaled, y_test)))
print("Number of non-zero features {}".format(np.sum(linridge.coef_ != 0)))
```

### Ridge Regression with various alpha parameter in combination with normalization.

```python
alphaarray =[0, 1, 10, 20, 50, 100, 1000]

for alpha in alphaarray:
    linridge = Ridge(alpha=alpha).fit(X_train_scaled, y_train)
    num_of_coeff_bigger = np.sum(linridge.coef_ > 1.0)
    print("Training Score: {:.3f} with alpha {}".format(linridge.score(X_train_scaled, y_train), alpha))
    print("Test Score: {:.3f} with aplha {}".format(linridge.score(X_test_scaled, y_test), alpha))
    print("Number of Absoulte coeffcient > 1.0 {} with alpha {}".format(num_of_coeff_bigger, alpha))
    
```

### Lasso Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_train, X_test, y_train, y_test = train_test_split(X_Crime, y_Crime, random_state = 0)
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

lassoReg = Lasso(alpha=20).fit(X_train_scaled, y_train)

print("Crime dataset")
print("Lasso regression linear model intercept: {}".format(lassoReg.intercept_))
print("Lasson regression linear model coefficient: {}".format(lassoReg.coef_))
print("R-Squared score (training) : {:.3f}".format(lassoReg.score(X_train_scaled, y_train)))
print("R-Squared score (test) : {:.3f}".format(lassoReg.score(X_test_scaled, y_test)))
print("Number of non-zero features {}".format(np.sum(lassoReg.coef_ != 0)))

## Plot feature which has non zero coef_.
for e in sorted(list(zip(list(X_Crime), lassoReg.coef_)), key = lambda e: -abs(e[1])):
     if e[1] != 0:
        print('\t{}, {:.3f}'.format(e[0], e[1]))
    


```

### Lasso Regression with regularization parameter

```python
alphaArray = [0.1, 1, 10, 50, 100]
x = range(len(alphaArray))
trainingScoreList = []
testScoreList = []
coeflist = []
for alpha in alphaArray:
    lassoReg = Lasso(alpha=alpha).fit(X_train_scaled, y_train)
    coeflist.append(np.mean(lassoReg.coef_))
    num_of_coeff_bigger = np.sum(lassoReg.coef_ != 0)
    trainingScore = lassoReg.score(X_train_scaled, y_train)
    print("Training Score: {:.3f} with alpha {}".format(trainingScore, alpha))
    trainingScoreList.append(trainingScore)
    testScore = lassoReg.score(X_test_scaled, y_test)
    print("Test Score: {:.3f} with aplha {}".format(testScore, alpha))
    print("Number of features kept {} with alpha {}".format(num_of_coeff_bigger, alpha))
    testScoreList.append(testScore)

plt.plot(x, trainingScoreList)
plt.plot(x, testScoreList)
# plt.plot(x, coeflist)
```

### Polynomial Regression
#### It leverages ridge regression but with multiple features..
#### Polynomial regression comes in handy if we want to co-relate 

```python
from sklearn.linear_model import Ridge
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_friedman1

X_F1, y_F1 = make_friedman1(n_samples = 100, n_features = 7, random_state = 0)
X_train, X_test, y_train, y_test = train_test_split(X_F1, y_F1, random_state = 0)

linReg = LinearRegression().fit(X_train, y_train)

print("Linear Regression .....")
print("Linear Reg Coef = {}".format(linReg.coef_))
print("Linear Reg Intercept = {:.2f}".format(linReg.intercept_))
print("Linear Regression R^2 training score {}".format(linReg.score(X_train, y_train)))
print("Linear Regression R^2 test score {}".format(linReg.score(X_test, y_test)))


# Convert X_F1 to polynomial feature with degree = 2
poly = PolynomialFeatures(degree=2)
X_F1_poly = poly.fit_transform(X_F1)
X_train_poly, X_test_poly, y_train, y_test = train_test_split(X_F1_poly, y_F1, random_state = 0)
linReg = LinearRegression().fit(X_train_poly, y_train)
print("Polynomial Regression .....")
print("Polynomial(deg=2) Coef = {}".format(linReg.coef_))
print("Polynomial(deg=2) Intercept = {:.2f}".format(linReg.intercept_))
print("Polynomial(deg=2) R^2 training score = {:.2f}".format(linReg.score(X_train_poly, y_train)))
print("Polynomial(deg=2) R^2 test score = {:.2f}".format(linReg.score(X_test_poly, y_test)))

# Lets utilize Ridge Regression with polynomial feature to see how scores looks like
ridReg = Ridge().fit(X_train_poly, y_train)
print("Ridge Regression .....")
print("(poly deg 2 + ridge) Coef = {}".format(ridReg.coef_))
print("(poly deg 2 + ridge) Intercept = {:.2f}".format(ridReg.intercept_))
print("(poly deg 2 + ridge) R^2 training score = {:.2f}".format(ridReg.score(X_train_poly, y_train)))
print("(poly deg 2 + ridge) R^2 test score = {:.2f}".format(ridReg.score(X_test_poly, y_test)))

# calculate Ridge Regression on polynomial feature with various alpha
alphaArray=[0.1, 1, 5, 10, 15, 20, 30, 50, 100]
for alpha, ax in alphaArray:
    ridReg = Ridge(alpha = alpha).fit(X_train_poly, y_train)
    print("Ridge Regression with normalization {}..............".format(alpha))
    print("\tRidge Regression with Polynomial(deg=2) Coef = {}".format(ridReg.coef_))
    print("\tRidge Regression with Polynomial(deg=2) Intercept = {:.2f}".format(ridReg.intercept_))
    print("\tRidge Regression with Polynomial(deg=2) R^2 training score = {:.2f}".format(ridReg.score(X_train_poly, y_train)))
    print("\tRidge Regression with Polynomial(deg=2) R^2 test score = {:.2f}".format(ridReg.score(X_test_poly, y_test)))
   
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from adspy_shared_utilities import (plot_class_regions_for_classifier_subplot)
from sklearn.model_selection import train_test_split
fig, subaxes = plt.subplots(1, 1, figsize=(7, 5))
y_fruit_apple = y_fruits_2d == 1 # make into a binary problem: apple vs everything else
X_train, X_test, y_train, y_test = train_test_split(X_fruits_2d.values, y_fruit_apple.values,
                                                   random_state = 0)
logisticReg = LogisticRegression(C=100).fit(X_train, y_train)
plot_class_regions_for_classifier_subplot(logisticReg, X_train, y_train, 
                                         None, None, 'Logistic Regression for binary classification.\n Fruits dataset: Apple vs Othes', subaxes)
h = 6
w = 8
print(logisticReg.predict([[h, w]]))
print('A fruit with height {} and width {} is predicted to be {}'.format(h, w, 
                                                                         ['not an apple', 'an apple']
                                                                         [logisticReg.predict([[h, w]])[0]]))

h=10
w=7
print(logisticReg.predict([[h, w]]))
print('A fruit with height {} and width {} is predicted to be {}'.format(h, w, 
                                                                         ['not an apple', 'an apple']
                                                                         [logisticReg.predict([[h, w]])[0]]))

subaxes.set_xlabel('height')
subaxes.set_ylabel('width')
print('Accuracy of Logistic regression classifier on training set: {:.2f}'
     .format(logisticReg.score(X_train, y_train)))
print('Accuracy of Logistic regression classifier on test set: {:.2f}'
     .format(logisticReg.score(X_test, y_test)))
```

```python
def answer_two():
    from sklearn.linear_model import LinearRegression
    from sklearn.preprocessing import PolynomialFeatures
    from sklearn.metrics.regression import r2_score
    from sklearn.model_selection import train_test_split
    degreeList = range(0, 10)
    predictedOutput = []
    list_r2_train = []
    list_r2_test = []
    for deg in degreeList:
        poly = PolynomialFeatures(degree = deg)
        X_train_poly = poly.fit_transform(X_train.reshape(11, 1))
        X_test_poly = poly.transform(X_test.reshape(4, 1))
        linReg = LinearRegression().fit(X_train_poly, y_train)
        y_train_predicted = linReg.predict(X_train_poly)
        y_test_predicted = linReg.predict(X_test_poly)
        r2_train = r2_score(y_train, y_train_predicted)
        r2_test = r2_score(y_test, y_test_predicted)
        list_r2_train.append(r2_train)
        list_r2_test.append(r2_test)
        np.asarray(list_r2_train), np.asarray(list_r2_test)
    return (np.asarray(list_r2_train), np.asarray(list_r2_test))
```
