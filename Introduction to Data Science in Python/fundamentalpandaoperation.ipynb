```python
#Broadcasting - applies to all values in the series
s = pd.Series(np.random.randint(0, 1000, 10000))
s.head()
```

```python
s += 2 #incrementing all values in the series by 2
s.head()
```

```python
#iterative way of incrementing the values in the series 
for label, value in s.iteritems():
    s.loc[label] = value + 2
s.head()
```

```python
for label, value in s.iteritems():
    s.at[label] = value + 2
s.head()
```
