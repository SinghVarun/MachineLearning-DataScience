```python
staff_df = pd.DataFrame([{'Name': 'Kelly', 'Role': 'Director of HR'},
                         {'Name': 'Sally', 'Role': 'Course liasion'},
                         {'Name': 'James', 'Role': 'Grader'}])
staff_df = staff_df.set_index('Name')
staff_df
```

```python
student_df = pd.DataFrame([{'Name': 'James', 'School': 'Business'},
                           {'Name': 'Mike', 'School': 'Law'},
                           {'Name': 'Sally', 'School': 'Engineering'}])
student_df = student_df.set_index('Name')
student_df
```

```python
# Merge operation (Union (outer join) and Intersection (inner join))
# Union/outer
pd.merge(student_df, staff_df, how='outer', left_index=True, right_index=True)
# since Kelly is only a staff member therefore she doesnt have a school assigned to her
# since Mike is only a student and doesn't participate in staff activities 
# there Role is NaN for Mike
```

```python
#intersection/inner join
pd.merge(student_df, staff_df, how='inner', left_index=True, right_index=True)
```

```python
# Get all staff whether they were student or not will come in next
pd.merge(staff_df, student_df, how='left', left_index=True, right_index=True)
```

```python
# Get all student whether they were staff or not will come in next
pd.merge(staff_df, student_df, how='right', left_index=True, right_index=True)
```

```python
# Merge method can also use columns to join on as seen below
pd.merge(staff_df, student_df, how='left', left_on='Name', right_on='Name')
```

```python
# Apply Operator in panda. 
# axis = 1 refers to all rows. 
# in short you basically apply a function on all rows in the dataframe.

import numpy as np
def min_max(row):
    data = row[['POPESTIMATE2010',
                'POPESTIMATE2011',
                'POPESTIMATE2012',
                'POPESTIMATE2013',
                'POPESTIMATE2014',
                'POPESTIMATE2015']]
    return pd.Series({'min': np.min(data), 'max': np.max(data)})

print(df.apply(min_max, axis=1))
```

```python
# now if we want to add new columns to the row 
def min_max(row):
    data = row[['POPESTIMATE2010',
                'POPESTIMATE2011',
                'POPESTIMATE2012',
                'POPESTIMATE2013',
                'POPESTIMATE2014',
                'POPESTIMATE2015']]
    row['min'] = np.min(data)
    row['max'] = np.max(data)
    return row


print(df.apply(min_max, axis=1))
```

```python
# using lamda function to compute max among selected columns 
rows = ['POPESTIMATE2010',
                'POPESTIMATE2011',
                'POPESTIMATE2012',
                'POPESTIMATE2013',
                'POPESTIMATE2014',
                'POPESTIMATE2015']

print(df.apply(lambda x: np.max(x[rows]), axis=1).head())
```

```python
def remove_digit_parenthesis(columnvalue):
    newcolumnvalue = ''.join(i for i in columnvalue if not i.isdigit())
    parenthesisindex = newcolumnvalue.find('(')
    if parenthesisindex > -1:
        newcolumnvalue = newcolumnvalue[:parenthesisindex]
    return newcolumnvalue.strip()


energryindicator['Country'] = energryindicator['Country'].apply(remove_digit_parenthesis)
```
