```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
```

# Subplot

```python
#create new figure
plt.figure()
# 1 row with 2 columns, and plot number = 1
plt.subplot(1, 2, 1) 

linear_data = np.array([1, 2, 3, 4, 5, 6])

plt.plot(linear_data, '-o')
```

```python
exponential_data = linear_data**2
# subplot with 1 rows, 2 columns and current axes is 2nd subplot axes
plt.subplot(1, 2, 2)
# do keep in mind that y axis is not shared between the two subplot. 
plt.plot(exponential_data, '-o')
```

```python
# plot exponential data on the first subplot itself. 
plt.subplot(1, 2, 1)
plt.plot(exponential_data, '-x')
```

```python
# make the two subplot share the axis
plt.figure()
ax1 = plt.subplot(1, 2, 1)
plt.plot(linear_data, '-o')
# Y axis between 2 plot is shared. 
ax2 = plt.subplot(1, 2, 2, sharey=ax1)
plt.plot(exponential_data, '-x')
```

```python
plt.figure()
# the right hand side is equivalent shorthand syntax
#1 * 100 + 2 * 10 + 1 * 1 (Hundered position, tens position, unit position)
plt.subplot(1,2,1) == plt.subplot(121)
```

```python
# create 3*3 grid of subplots (remember its subplot(s) and not subplot). 

fig, ((ax1, ax2, ax3), (ax4, ax5, ax6), (ax7, ax8, ax9)) = plt.subplots(3, 3, sharex=True, sharey=True)

# Plot linear data on any subplot, for e.g. we choosed 2nd subplot to have the linear data
ax2.plot(linear_data, '-x')
```

```python
## set inside tick labels (for x and y axis) to visible
for ax in plt.gcf().get_axes():
    for label in ax.get_xticklabels() + ax.get_yticklabels():
        label.set_visible = True
```

# Histograms

```python
# create 2 * 2 grid of axis of subplots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex=True)
ax = [ax1, ax2, ax3, ax4]

for n in range(0, len(ax)):
    sample_size = 10**(n+1)
    sample = np.random.normal(loc=0, scale=1.0, size=sample_size)
    ax[n].hist(sample)
    ax[n].set_title("n={}".format(sample_size))

```

```python
# lets sets the bin size to 100 for above histogram
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex=True)
ax = [ax1, ax2, ax3, ax4]

for n in range(0, len(ax)):
    sample_size = 10**(n+1)
    sample = np.random.normal(loc=0, scale=1.0, size=sample_size)
    ax[n].hist(sample, bins=100)
    ax[n].set_title("n={}".format(sample_size))
```

### gridspec

```python
# A class that speicifies geometery of grid that a subplot will be placed. Location of grid is determined in similar way as Subplotparams
import matplotlib.gridspec as gridspec

plt.figure()
# Gridspec takes in number of rows, number of columns. 
gspec = gridspec.GridSpec(3, 3)

top_histogram = plt.subplot(gspec[0, 1:]) # gpec gives the co-ordinates. In this case its first rows starting at 2nd Column and extending till the end.
side_histogram = plt.subplot(gspec[1:, 0]) # gpec gives the co-ordinates. In this case its 2nd rows and continue till end of all rows, constrained to only 3rd Column.
lower_right = plt.subplot(gspec[1:, 1:]) # gpec gives the co-ordinates. In this case it starts at 2nd rows and continue till end of all rows, and expands from 2nd column onwards till  3rd Column.

Y = np.random.normal(loc=0.0, scale=1.0, size=10000)
X = np.random.random(size=10000)

lower_right.scatter(X, Y)
top_histogram.hist(X, bins=100)
s = side_histogram.hist(Y, bins=100, orientation='horizontal')
```

```python
# lets normalize the histogram.
plt.figure()
# Gridspec takes in number of rows, number of columns. 
gspec = gridspec.GridSpec(3, 3)

top_histogram = plt.subplot(gspec[0, 1:]) # gpec gives the co-ordinates. In this case its first rows starting at 2nd Column and extending till the end.
side_histogram = plt.subplot(gspec[1:, 0]) # gpec gives the co-ordinates. In this case its 2nd rows and continue till end of all rows, constrained to only 3rd Column.
lower_right = plt.subplot(gspec[1:, 1:]) # gpec gives the co-ordinates. In this case it starts at 2nd rows and continue till end of all rows, and expands from 2nd column onwards till  3rd Column.

Y = np.random.normal(loc=0.0, scale=1.0, size=10000)
X = np.random.random(size=10000)


top_histogram.hist(X, bins=100, density=True)
side_histogram.hist(Y, bins=100, orientation='horizontal', density=True)
side_histogram.invert_xaxis()
lower_right.scatter(X, Y)
```

```python
for ax in [top_histogram, lower_right]:
    ax.set_xlim(0, 1)

for ax in [side_histogram, lower_right]:
    ax.set_ylim(-5, 5)
```

# Box and Whisker Plots

```python
# Box plot is a method of showing aggregate statistics of various samples in concise manner
# Box plot simultaneously shows, for each sample median of each, the minimum and maximum of the samples and the interquartile range
# Interquartile range (0-25%, 25%-50%, 50%-75%, 75%-100%)
import numpy as np
import pandas as pd
normal_sample = np.random.normal(loc=0.0, scale=1.0, size=10000)
random_sample = np.random.random(size=10000)
gamma_sample = np.random.gamma(2, size=10000)

# Create a dataframe as most the values that we are going to process will come from panda

df = pd.DataFrame({"normal": normal_sample, 
                  "random": random_sample,
                  "gamma": gamma_sample})
# Percentage value 25%, 50%, 75% describes the interquartile range.
df.describe()
```

```python
plt.figure()
#create a boxplot of the normal data, assign the output to a variable (_) to supress the output, as we dont require to use that
# variable
_ = plt.boxplot(df['normal'], whis='range')
```

```python
# clear the current figure
plt.clf()

_ = plt.boxplot([df['normal'], df['random'], df['gamma']], whis='range')
```

```python
plt.figure()
_ = plt.hist(df['gamma'], bins=100)
```

```python
# We can actually overlay an axes on top of another within a figure
# This functionality isn't in the basic matplotlib space, but it is in the toolkits which tends to ship with matplotlib. 
# we import toolkit axes_grid 
import mpl_toolkits.axes_grid1.inset_locator as mpl_il

plt.figure()
_ = plt.boxplot([df['normal'], df['random'], df['gamma']], whis='range')
ax2= mpl_il.inset_axes(plt.gca(), width='60%', height='40%', loc=2)
ax2.hist(df['gamma'], bins=100)
ax2.margins(x=0.5)
```

```python
ax2.yaxis.tick_right()
```

```python
plt.figure()
_ = plt.boxplot([df['normal'], df['random'], df['gamma']], whis='range')
```

```python
# First, if you don't supply the whis argument, the whiskers actually only go out to halfway between the interquartile range.
plt.figure()
_ = plt.boxplot([df['normal'], df['random'], df['gamma']])
```
