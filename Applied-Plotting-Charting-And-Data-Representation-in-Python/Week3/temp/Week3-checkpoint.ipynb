```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
```

```python
#create new figure
plt.figure()
# 1 row with 2 columns, and plot number = 1
plt.subplot(1, 2, 1) 

linear_data = np.array([1, 2, 3, 4, 5, 6])

plt.plot(linear_data, '-o')
```

```python
exponential_data = linear_data**2
# subplot with 1 rows, 2 columns and current axes is 2nd subplot axes
plt.subplot(1, 2, 2)
# do keep in mind that y axis is not shared between the two subplot. 
plt.plot(exponential_data, '-o')
```

```python
# plot exponential data on the first subplot itself. 
plt.subplot(1, 2, 1)
plt.plot(exponential_data, '-x')
```

```python
# make the two subplot share the axis
plt.figure()
ax1 = plt.subplot(1, 2, 1)
plt.plot(linear_data, '-o')
# Y axis between 2 plot is shared. 
ax2 = plt.subplot(1, 2, 2, sharey=ax1)
plt.plot(exponential_data, '-x')
```

```python
plt.figure()
# the right hand side is equivalent shorthand syntax
#1 * 100 + 2 * 10 + 1 * 1 (Hundered position, tens position, unit position)
plt.subplot(1,2,1) == plt.subplot(121)
```

```python
# create 3*3 grid of subplots (remember its subplot(s) and not subplot). 

fig, ((ax1, ax2, ax3), (ax4, ax5, ax6), (ax7, ax8, ax9)) = plt.subplots(3, 3, sharex=True, sharey=True)

# Plot linear data on any subplot, for e.g. we choosed 2nd subplot to have the linear data
ax2.plot(linear_data, '-x')
```

```python
## set inside tick labels (for x and y axis) to visible
for ax in plt.gcf().get_axes():
    for label in ax.get_xticklabels() + ax.get_yticklabels():
        label.set_visible=True
```
