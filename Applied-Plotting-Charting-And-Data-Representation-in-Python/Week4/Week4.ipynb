```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib notebook
```

```python
np.random.seed(1234)

v1 = pd.Series(np.random.normal(0, 10, 1000), name='v1')
v2 = pd.Series(2 * v1 + np.random.normal(60, 15, 1000), name='v2')
plt.figure()
plt.hist(v1, alpha=0.7, bins=np.arange(-50, 150, 5), label='v1');
plt.hist(v2, alpha=0.7, bins=np.arange(-50, 150, 5), label='v2');
plt.legend();
```

```python
plt.figure();
plt.hist([v1, v2], histtype='barstacked', density=True);
v3 = np.concatenate((v1, v2))
sns.kdeplot(v3)
```

```python
plt.figure()
sns.distplot(v3, hist_kws={'color': 'Teal'}, kde_kws={'color':'Navy'});
```

```python
# The jointplot creates a scatterplot along the histograms for each individual variable on each axis. 
#Setting alpha to 0.4 will help with visualizing overlapping points. The jointplot allows us to visualize the distribution of the two variables individually. As well as relationships between the variables
plt.figure();
sns.jointplot(v1, v2, alpha=0.4);
```

```python
plt.figure();
grid = sns.jointplot(v1, v2, alpha=0.4);
grid.ax_joint.set_aspect('equal');
```

```python
# Hexbin plots are the bivariate counterpart to histograms. 
# Hexbin plots show the number of observations that fall within hexagonal bins. 
sns.jointplot(v1, v2, kind='hex');
```

```python
## Set Seaborn style to white, so all plot coming after this will have there grey tile turned off.
sns.set_style('white')

sns.jointplot(v1, v2, kind='kde');
```

```python
iris = pd.read_csv('iris.csv')
iris.head()
```

```python
# similar to scatterplot_matrix in matplotlib, seaborn has a method which does the same
sns.pairplot(iris, hue='variety', diag_kind='kde', size=2)
```

```python
# violin plot is an informative version of a boxplot. 
# swarm plot is a scatter plot for categorical data
# Looking at the swarmplot, each species has its own column and each observation's petal length is shown. 
# With more common values appearing as the wide parts of the cluster, much like a histogram. 
# The violinplot is like box plot with a rotated kernel density estimation on each side. 
# Violin plots convey more information than box plots and are able to show particular phenomena within the distribution 
# that box plots are unable to convey such as multi modality.
plt.figure(figsize=(8, 6))
plt.subplot(121)
sns.swarmplot('variety', 'sepal.length', data=iris);
plt.subplot(122)
sns.violinplot('variety', 'sepal.length', data=iris);
```
