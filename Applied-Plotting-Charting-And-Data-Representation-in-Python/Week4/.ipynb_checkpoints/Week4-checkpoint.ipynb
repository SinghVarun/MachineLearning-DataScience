```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(123)
df = pd.DataFrame({'A': np.random.randn(365).cumsum(),
                  'B': np.random.randn(365).cumsum() + 20,
                  'C': np.random.randn(365).cumsum() - 20}, index=pd.date_range('1/1/2017', periods=365))

df.plot('A', 'C', kind='scatter')
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-colorblind')
```

```python
df.plot()
```

```python
df.plot.scatter('A', 'C', c='B', s=df['B'], colormap='viridis')
```

```python
ax = df.plot.scatter('A', 'C', c='B', s=df['B'], colormap='viridis');
ax.set_aspect('equal')
```

```python
df.plot.box();
```

```python
df.plot.hist();
```

```python
df.plot.kde();
```
